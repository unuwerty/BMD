<h3>Візуалізація даних за допомогою бібліотеки Seaborn</h3>


```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

<h6>Завантажуємо данні з вбудованих баз данних</h6>

```python
ir = sns.load_dataset('iris')
```

```python
ir.head()
```

```python
sns.jointplot(x='sepal_length', y='sepal_width', data = ir)
```

```python
sns.distplot(ir['sepal_width'], kde=False)
```

```python
sns.boxplot(x='sepal_width',y='sepal_length', data=ir, palette="rainbow")
```

```python
import numpy as np
sns.barplot(x='sepal_width', y='sepal_length', data=ir,estimator=np.median)
```

```python
sns.stripplot(x='sepal_width', y='sepal_length', data=ir)
```

```python
sns.heatmap(ir.corr())
```

```python
def sinplot(flip=1):
    x = np.linspace(0, 10, 50)
    for i in range(1, 5):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
sinplot()
```

```python
g = sns.FacetGrid(ir, col="species")
g = g.map(plt.hist, "sepal_length")
```
