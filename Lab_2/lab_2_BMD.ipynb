<h3>Імпортування</h3>

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

<h3>Основи побудови графіків</h3>

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

<h3>Побудова графіків</h3>

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y label')
plt.title('Title')
```

<h6>Для побудови декількох графіків поруч, використовуйте команду
subplot:</h6>

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

<h6>Об’єктно-орієнтований метод</h6>

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Title')
```

<h6>Для одночасної побудови декількох графіків додайте їх до об’єкта з
вказанням області відображення:</h6>

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.1,0.5,0.4,0.3])

axes1.plot(x,y,'r')
axes1.set_xlabel('X label')
axes1.set_ylabel('Y label')
axes1.set_title('Title')

axes2.plot(x,y,'b')
axes2.set_xlabel('X label')
axes2.set_ylabel('Y label')
axes2.set_title('Title')
```

<h6>Для використання функції subplots при створенні графіка як об’єкта,
його ініціалізація має бути виконана наступним чином:</h6>

```python
fig, axes = plt.subplots(nrows=1,ncols=3)

axes[0].plot(x,y)
axes[1].plot(y,x)
axes[2].plot(x,x)
```

<h6>Використання команди .legend()</h6>

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2, label="x**2")
ax.plot(x,x**3, label="x**3")
ax.plot(x,x**4, label="x**4")
ax.plot(x,x**0.5, label="x**0.5")
ax.legend()
```

<h6>Зміна кольору графіків</h6>

```python
fig, ax = plt.subplots()
ax.plot(x,x**2,color = 'y')
ax.plot(x,x**3,color = 'r')
```

<h6>Зміна типу та ширини ліній</h6>

```python
fig, ax = plt.subplots()
ax.plot(x,x**2,lw=2,ls = '--')
```

<h6>Зміна типу, розміру, та кольору маркерів</h6>

```python
fig, ax = plt.subplots()
ax.plot(x,x**2,marker = 'o', markersize = 10, markerfacecolor = 'black')
```
